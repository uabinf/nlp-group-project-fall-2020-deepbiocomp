```python
rename = {'QType' : 'title',
'Question' : 'question', 
'Answer': 'context'}
```

```python
import pandas as pd
import nltk
import json
```

```python
df = pd.read_csv("dev_qa.tsv", sep="\t")
df.rename(columns=rename, inplace=True)
cols = ["title", 'question', 'context']
df = df[cols]
lol = df.values.tolist()
cols += ["text", "answer_start", "id"]
LOL = []
ID = 0
for i in lol:
    Answer = nltk.sent_tokenize(i[-1])[0]
    answer_start = 0
    i.append(Answer)
    i.append(answer_start)
    ID += 1
    i.append(str(ID))
    LOL.append(i)
    #print(i)
    #break
del(df)
df = pd.DataFrame(LOL, columns=cols)
df

#print(LOL[0])
i = 0

data = []

for i in range(len(LOL)):
    data_dic =  {
            "title":LOL[i][0], 
            "paragraphs":[
                {"context":LOL[i][2], 
                 "qas":[
                     {"answers":
                      [
                          {"answer_start": LOL[i][4],
                          "text":LOL[i][3]
                          }
                      ], 
                      "question":LOL[i][1],
                      "id":LOL[i][5]
                     }
                 ]
                }
            ]
        }
    data.append(data_dic)
Data = {"data" : data,
       'version' : "1.1",
       "team" : "nlp-group-project-fall-2020-deepbiocomp"}

#print(Data)
json.dumps(Data, indent = 4) 

with open('dev_qa.json', 'w') as fp:
    json.dump(Data, fp)
```
